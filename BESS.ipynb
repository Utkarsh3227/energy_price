```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from darts import TimeSeries
from darts.models import LinearRegressionModel
from darts.metrics import mae, rmse
from sklearn.preprocessing import StandardScaler

# Assuming month_df is already loaded and contains the 'RAISE5MIN' column
month_df = month_df.sort_index()
# Fill missing dates and forward fill missing values
month_df = month_df.asfreq('5T', method='ffill')

# Convert the DataFrame to a TimeSeries
series = TimeSeries.from_dataframe(month_df, value_cols='RAISE5MIN', fill_missing_dates=True, freq='5T')

# Define the number of test steps (e.g., 3 days of data at 5-minute intervals)
n_test_steps = 12 * 24 * 3  # Number of 5-minute intervals in the test set (3 days)

# Split the data into training and testing sets
train, test = series[:-n_test_steps], series[-n_test_steps:]

# Define and train the linear regression model
model = LinearRegressionModel(lags=12*24*7)  # Lags for one week
model.fit(train)

# Forecast the values for the test set period
forecast = model.predict(n=n_test_steps)

# Calculate the performance metrics
mae_val = mae(test, forecast)
rmse_val = rmse(test, forecast)

print(f'Mean Absolute Error (MAE): {mae_val}')
print(f'Root Mean Squared Error (RMSE): {rmse_val}')

# Plot the observed and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(train.time_index, train.values().flatten(), label='Training')
plt.plot(test.time_index, test.values().flatten(), label='Test', color='blue')
plt.plot(forecast.time_index, forecast.values().flatten(), label='Forecast', color='orange')
plt.title('RAISE5MIN Forecast vs Actuals (Scaled)')
plt.xlabel('Date Time')
plt.ylabel('Scaled RAISE5MIN')
plt.legend()
plt.show()

# Calculate mean of non-outliers
mean_non_outliers = month_df['RAISE5MIN'].mean()

# Plot the results of forecasting by splitting the graph weekly for better visualization
start_date = month_df.index.min()
end_date = month_df.index.max()

current_start_date = start_date
one_week = pd.DateOffset(weeks=1)

while current_start_date < end_date:
    current_end_date = current_start_date + one_week
    mask = (month_df.index >= current_start_date) & (month_df.index < current_end_date)
    subset_df = month_df.loc[mask]

    if not subset_df.empty:
        # Plot the subset
        plt.figure(figsize=(12, 6))
        plt.plot(subset_df.index, subset_df['RAISE5MIN'], label='RAISE5MIN (Scaled)')

        # Plot forecast values if they fall within the current week
        forecast_mask = (forecast.time_index >= current_start_date) & (forecast.time_index < current_end_date)
        if forecast_mask.sum() > 0:
            plt.plot(forecast.time_index[forecast_mask], forecast.values().flatten()[forecast_mask], label='Forecast', color='orange')

        plt.axhline(mean_non_outliers, color='r', linestyle='--', label='Mean of Non-Outliers')
        plt.legend()
        plt.xlabel('Date Time')
        plt.ylabel('Scaled RAISE5MIN')
        plt.title(f'Time Series Data from {current_start_date.strftime("%Y-%m-%d")} to {current_end_date.strftime("%Y-%m-%d")}')
        plt.show()

    current_start_date = current_end_date
```

```python
import numpy as np
import pandas as pd
from omnixai.data.timeseries import Timeseries
from omnixai.explainers.timeseries import TimeseriesExplainer
import plotly.io as pio

# Set Plotly renderer to display in Jupyter
pio.renderers.default = "png"

# Load your time series dataset
df = month_df[["RAISE5MIN"]]  # Change the column as needed
print(df.head())

# Split the dataset into training and test sets (80% train, 20% test)
train_df = df.iloc[:int(len(df)*0.8)]
test_df = df.iloc[int(len(df)*0.8):]

# Check the size of train and test sets
print(f"Training set size: {len(train_df)}")
print(f"Test set size: {len(test_df)}")

# Define a threshold for anomaly detection (e.g., 90th percentile of training data)
threshold = np.percentile(train_df["RAISE5MIN"].values, 90)

# Define a simple detector function to identify anomalies
def detector(ts: Timeseries):
    # Calculate the anomaly score based on the threshold
    anomaly_scores = np.sum((ts.values > threshold).astype(int))
    return anomaly_scores / ts.shape[0]

# Convert the training data to Timeseries format
train_timeseries = Timeseries.from_pd(train_df)

# Initialize a TimeseriesExplainer with SHAP and MACE for anomaly detection
explainers = TimeseriesExplainer(
    explainers=["shap", "mace"],
    mode="anomaly_detection",
    data=train_timeseries,
    model=detector,
    preprocess=None,
    postprocess=None,
    params={
        "mace": {
            "threshold": 0.001,
            "num_instances": min(5, len(test_df))  # Adjusted number of instances for MACE
        }
    }
)

# Convert the test data to Timeseries format
test_timeseries = Timeseries.from_pd(test_df)

# Generate explanations with error handling
local_explanations = None  # Initialize to avoid NameError
try:
    local_explanations = explainers.explain(
        test_timeseries,
        params={"shap": {"nsamples": 500}}  # Reduced number of samples for SHAP
    )
except IndexError as e:
    print(f"Error during explanation: {e}")
except Exception as e:
    print(f"Unexpected error during explanation: {e}")

# Check if explanations were generated successfully
if local_explanations is not None:
    # Ensure the selected index is within bounds
    index = min(0, len(test_timeseries) - 1)

    # Plot SHAP results
    print("SHAP results:")
    try:
        local_explanations["shap"].ipython_plot(index)
    except IndexError as e:
        print(f"Error in SHAP plot: {e}")
    except Exception as e:
        print(f"Unexpected error in SHAP plot: {e}")

    # Plot MACE results
    print("MACE results:")
    try:
        local_explanations["mace"].ipython_plot(index)
    except IndexError as e:
        print(f"Error in MACE plot: {e}")
    except Exception as e:
        print(f"Unexpected error in MACE plot: {e}")

    # Launch the dashboard to visualize the explanations interactively
    # dashboard = Dashboard(instances=test_timeseries, local_explanations=local_explanations)
    # dashboard.show()
else:
    print("No explanations generated due to previous errors.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from darts import TimeSeries
from darts.models import Prophet
from darts.metrics import mae, rmse

# Fill missing dates and forward fill missing values
df = month_df.asfreq('5T', method='ffill')

# Create additional features (e.g., lagged features)
df['external_regressor'] = df['RAISE5MIN'].shift(1).fillna(0)

# Convert the DataFrame to TimeSeries
series = TimeSeries.from_dataframe(df, value_cols='RAISE5MIN', fill_missing_dates=True, freq='5T')
regressor_series = TimeSeries.from_dataframe(df, value_cols='external_regressor', fill_missing_dates=True, freq='5T')

# Define the number of test steps for the last three days
n_test_steps = 12 * 24 * 3  # Number of 5-minute intervals in 3 days

# Split the data into training and testing sets
train, test = series[:-n_test_steps], series[-n_test_steps:]
regressor_train, regressor_test = regressor_series[:-n_test_steps], regressor_series[-n_test_steps:]

# Use Prophet to fit the model with external regressors
model = Prophet()
model.fit(train, future_covariates=regressor_train)

# Forecast the values for the test set period
forecast = model.predict(n=n_test_steps, future_covariates=regressor_test)

# Calculate the performance metrics
mae_val = mae(test, forecast)
rmse_val = rmse(test, forecast)

print(f'Mean Absolute Error (MAE): {mae_val}')
print(f'Root Mean Squared Error (RMSE): {rmse_val}')

# Plot the observed and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(train.time_index, train.values().flatten(), label='Training')
plt.plot(test.time_index, test.values().flatten(), label='Test', color='blue')
plt.plot(forecast.time_index, forecast.values().flatten(), label='Forecast', color='orange')
plt.title('Prophet Forecast with External Regressors')
plt.xlabel('Date Time')
plt.ylabel('RAISE5MIN')
plt.legend()
plt.show()

# Zoom in on the last three days of the test set for detailed analysis
last_three_days = pd.DateOffset(days=3)
last_three_days_start = test.time_index[-1] - last_three_days
last_three_days_end = test.time_index[-1]

test_mask = (test.time_index >= last_three_days_start) & (test.time_index <= last_three_days_end)
forecast_mask = (forecast.time_index >= last_three_days_start) & (forecast.time_index <= last_three_days_end)

# Plot the forecast and actual values for the last three days
plt.figure(figsize=(12, 6))
plt.plot(test.time_index[test_mask], test.values().flatten()[test_mask], label='Test (Last 3 Days)', color='blue')
plt.plot(forecast.time_index[forecast_mask], forecast.values().flatten()[forecast_mask], label='Forecast (Last 3 Days)', color='orange')

plt.legend()
plt.xlabel('Date Time')
plt.ylabel('RAISE5MIN')
plt.title('Forecast vs Actuals for the Last 3 Days')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor
import shap

# Function to create lagged features
def create_lagged_features(df, lags):
    df_lagged = df.copy()
    for lag in range(1, lags + 1):
        df_lagged[f'lag_{lag}'] = df_lagged['RAISE5MIN'].shift(lag)
    df_lagged = df_lagged.dropna()
    return df_lagged

# Create lagged features
lags = 12*24  # Number of lags to include (adjust as needed)
df_lagged = create_lagged_features(month_df, lags)

# Remove non-numeric columns (e.g., 'date_time') if present
df_lagged = df_lagged.drop(columns=['time'], errors='ignore')

# Define features and target
X = df_lagged.drop('RAISE5MIN', axis=1)
y = df_lagged['RAISE5MIN']

# Define the number of test steps (adjust as needed)
n_test_steps = 24*12 * 3  # 48 lags * 10 = 480, which corresponds to 480 * 5 min = 2400 min (40 hours)

# Split the data into training and testing sets
X_train, X_test = X[:-n_test_steps], X[-n_test_steps:]
y_train, y_test = y[:-n_test_steps], y[-n_test_steps:]

# Initialize and fit the XGBoost model
model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6)
model.fit(X_train, y_train)

# Forecast the values for the test set period
y_pred = model.predict(X_test)

# Calculate the performance metrics
mae_val = mean_absolute_error(y_test, y_pred)
rmse_val = np.sqrt(mean_squared_error(y_test, y_pred))

print(f'Mean Absolute Error (MAE): {mae_val}')
print(f'Root Mean Squared Error (RMSE): {rmse_val}')

# Plot the observed and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(month_df.index[-len(y_test):], y_test, label='Test', color='blue')
plt.plot(month_df.index[-len(y_test):], y_pred, label='Forecast', color='orange')
plt.title('XGBoost Forecast vs Actuals')
plt.xlabel('Date Time')
plt.ylabel('RAISE5MIN')
plt.legend()
plt.show()

# --- SHAP Explanation (Independent of AIX360) ---

# Using SHAP for Global Feature Importance
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)

# Plot SHAP summary for global feature importance
shap.summary_plot(shap_values, X_test, plot_type="bar")

# SHAP Force Plot (explaining one prediction)
shap.force_plot(explainer.expected_value, shap_values[0, :], X_test.iloc[0, :], matplotlib=True)
```
