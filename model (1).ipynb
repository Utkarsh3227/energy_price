```python
df = df1[df1['REGIONID'] == 'NSW1']
df.head(10)
```

```python
df = df[['PERIODID','date_time','RRP']]
```

```python
df.index = df.index//5 + 1
```

```python
df['time'] = pd.to_datetime(df['date_time'])
```

```python
df = df[df['time'].dt.minute % 30 == 0]

df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Calculate Z-scores

df['z_score'] = (df['RRP'] - df['RRP'].mean()) / df['RRP'].std()

# Define a threshold for Z-score
threshold = 2

# Calculate the mean of non-outlier data
mean_non_outliers = df[np.abs(df['z_score']) < threshold]['RRP'].mean()

# Replace outliers with the mean
df.loc[np.abs(df['z_score']) >= threshold, 'RRP'] = mean_non_outliers

# Drop the Z-score column as it's no longer needed
df.drop(columns=['z_score'], inplace=True)

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['RRP'], label='RRP')
plt.axhline(mean_non_outliers, color='r', linestyle='--', label='Mean of Non-Outliers')
plt.legend()
plt.xlabel('Date Time')
plt.ylabel('RRP')
plt.title('Time Series Data with Outliers Replaced by Mean')
plt.show()
```

```python
start_date = pd.to_datetime('2015-01-01')
end_date = pd.to_datetime('2015-01-31')

current_start_date = start_date
one_week = pd.DateOffset(weeks=1)

while current_start_date < end_date:
    current_end_date = current_start_date + one_week
    mask = (df['time'] >= current_start_date) & (df['time'] < current_end_date)
    subset_df = df.loc[mask]

    # Plot the subset
    plt.figure(figsize=(12, 6))
    plt.plot(subset_df['time'], subset_df['RRP'], label='RRP')
    plt.axhline(mean_non_outliers, color='r', linestyle='--', label='Mean of Non-Outliers')
    plt.legend()
    plt.xlabel('Date Time')
    plt.ylabel('RRP')
    plt.title(f'Time Series Data from {current_start_date.strftime("%Y-%m-%d")} to {current_end_date.strftime("%Y-%m-%d")}')
    plt.show()

    current_start_date = current_end_date
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
month_df
```

```python

month_df['time'] = pd.to_datetime(month_df['time'])
month_df.set_index('time', inplace=True)
```

```python
month_df
```

month_df

```python
n_test_steps = 48 * 2  # Number of 30-minute intervals in the test set


# Split the data into training and testing sets
train, test = month_df[:-n_test_steps], month_df[-n_test_steps:]

# Define the SARIMA model with identified parameters
p, d, q = 1, 1, 1
P, D, Q, s = 1, 1, 1, 48

# Fit the SARIMA model on the training set
model = SARIMAX(train['RRP'], order=(p, d, q), seasonal_order=(P, D, Q, s))
model_fit = model.fit(disp=False)

# Forecast the values for the test set period
forecast = model_fit.get_forecast(steps=n_test_steps)
forecast_index = pd.date_range(start=train.index[-1], periods=n_test_steps + 1, freq='30T')[1:]

# Calculate the performance metrics
y_true = test['RRP']
y_pred = forecast.predicted_mean

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Plot the observed and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['RRP'], label='Training')
plt.plot(test.index, test['RRP'], label='Test', color='blue')
plt.plot(forecast_index, y_pred, label='Forecast', color='orange')
plt.fill_between(forecast_index, forecast.conf_int().iloc[:, 0], forecast.conf_int().iloc[:, 1], color='orange', alpha=0.3)
plt.title('RRP Forecast vs Actuals')
plt.xlabel('Date Time')
plt.ylabel('RRP')
plt.legend()
plt.show()
```
