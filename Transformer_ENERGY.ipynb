```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/AllPrices15_24/AllPrices15_24.csv')
df1.head(10)
```

```python
df = df1[df1['REGIONID'] == 'NSW1']
df.head(10)
```

```python
df = df[['PERIODID','date_time','RRP']]
df
```

```python
df.index = df.index//5 + 1
df['time'] = pd.to_datetime(df['date_time'])

df = df[df['time'].dt.minute % 30 == 0]

df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate Z-scores

df['z_score'] = (df['RRP'] - df['RRP'].mean()) / df['RRP'].std()

# Define a threshold for Z-score
threshold = 2

# Calculate the mean of non-outlier data
mean_non_outliers = df[np.abs(df['z_score']) < threshold]['RRP'].mean()

# Replace outliers with the mean
df.loc[np.abs(df['z_score']) >= threshold, 'RRP'] = mean_non_outliers

# Drop the Z-score column as it's no longer needed
df.drop(columns=['z_score'], inplace=True)

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['RRP'], label='RRP')
plt.axhline(mean_non_outliers, color='r', linestyle='--', label='Mean of Non-Outliers')
plt.legend()
plt.xlabel('Date Time')
plt.ylabel('RRP')
plt.title('Time Series Data with Outliers Replaced by Mean')
plt.show()
```

```python
start_date = pd.to_datetime('2019-01-01')
end_date = pd.to_datetime('2024-02-21')

# Filter the DataFrame for the desired date range
# month_df = df[(df['time'] >= start_date) & (df['time'] <= end_date)]
long_df = df[(df['time'] >= start_date) & (df['time'] <= end_date)]
# ten_day_df = df[(df['time'] >= start_date) & (df['time'] <= end_date)]
# ten_day_df
long_df
```

```python

long_df['time'] = pd.to_datetime(long_df['time'])
long_df.set_index('time', inplace=True)

long_df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
# month_df['RRP'] = scaler.fit_transform(month_df[['RRP']])
long_df['RRP'] = scaler.fit_transform(long_df[['RRP']])

long_df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from darts import TimeSeries
from darts.models import TransformerModel
from darts.metrics import mae, rmse, r2_score, mse

# Load your dataset and handle missing values
df = long_df.asfreq('30min', method='ffill')

# Convert the DataFrame to a TimeSeries
series = TimeSeries.from_dataframe(df, value_cols='RRP', fill_missing_dates=True, freq='30min')

# Define the number of steps for the test set
n_test_steps = 48 * 30 * 6  # 6 weeks of 30-minute intervals

# Split the data into training and testing sets
train, test = series[:-n_test_steps], series[-n_test_steps:]

# Use Transformer model to fit the data
model = TransformerModel(
    input_chunk_length=48,
    output_chunk_length=2,
    d_model=64,
    nhead=4,
    num_encoder_layers=3,
    num_decoder_layers=3,
    dropout=0.1,
    n_epochs=30,
    random_state=42,
    optimizer_kwargs={'lr': 0.001},  # Set learning rate here
)

model.fit(train)

# Forecast the values for the test set period
forecast = model.predict(n_test_steps)

# Calculate the performance metrics
mae_val = mae(test, forecast)
rmse_val = rmse(test, forecast)
r2_val = r2_score(test, forecast)
mse_val = mse(test, forecast)

print(f"Mean Absolute Error (MAE): {mae_val}")
print(f"Root Mean Squared Error (RMSE): {rmse_val}")
print(f"RÂ² Score: {r2_val}")
print(f"Mean Squared Error (MSE): {mse_val}")

# Plot the observed and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(train.time_index, train.values().flatten(), label='Training')
plt.plot(test.time_index, test.values().flatten(), label='Test', color='blue')
plt.plot(forecast.time_index, forecast.values().flatten(), label='Forecast', color='orange')
plt.title('RRP Forecast vs Actuals')
plt.xlabel('Date Time')
plt.ylabel('Scaled RRP')
plt.legend()
plt.show()
```
